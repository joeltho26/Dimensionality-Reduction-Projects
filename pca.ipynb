```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import  pandas as pd
from sklearn.datasets import load_breast_cancer
%matplotlib inline
```

```python
cancer = load_breast_cancer()
cancer.keys()
```

```python
input_data = cancer['data']
labels = cancer['target']
```

```python
data = pd.DataFrame(input_data, columns=cancer['feature_names'])
data.head(5)
```

```python
data.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
scaled_data
```

```python
labels.shape
```

```python
from sklearn.decomposition  import PCA
pca = PCA(n_components=2)
final_data = pca.fit_transform(scaled_data)
len(final_data)
```

```python
final_data.shape
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(final_data[:,0], final_data[:,1], c=labels)
plt.xlabel("Principle Component Analysis 1")
plt.ylabel("Principle Component Analysis 2")
plt.show()
```
